# import lib

````python
import cv2
import matplotlib.pyplot as plt
````

# read and display img

````python
path_img = "./img/download.jpg"
````

````python
img = cv2.imread(path_img)
img 
# ''' 
# ảnh sẽ là một ma trận 
# '''
````

````python

print(img.shape)
# '''
#     khi .shape thì sẽ hiện kích thước ảnh là :
#     Height , Width , Channel
# '''
````

````python
H, W, C = img.shape
print("Height: ", H)
print("Width: ", W)
print("Channel: ", C)
````

````python
print(img.size) # size of the image = H * W * C
print("Size: ", H*W*C)
# '''
#     khi .size thì sẽ sẽ số pixel trong ảnh
# '''
````

````python
img.dtype
# '''
#     Kiểu dữ liệu của ảnh
# '''
````

````python
plt.imshow(img) # ảnh sẽ hiện ra khác với màu gốc vì cv2 đọc ảnh theo BGR 
plt.show()
````

````python
plt.imshow(img[:,:,::-1]) 
plt.show()      
````

````python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
````
