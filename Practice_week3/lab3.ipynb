### Biến Đổi Luật Lũy Thừa: 
- Bao gồm các phép biến đổi lũy thừa bậc n và căn bậc n. Những phép biến đổi này có thể được biểu diễn bằng biểu thức:
- s=cr^γ
 
- Ký hiệu γ được gọi là gamma, vì thế phép biến đổi này cũng được biết đến với tên gọi biến đổi gamma.

- Sự thay đổi giá trị của γ làm thay đổi mức độ tăng cường của hình ảnh. Các thiết bị hiển thị khác nhau / màn hình có mức hiệu chỉnh gamma riêng, vì vậy chúng hiển thị hình ảnh với cường độ khác nhau.

- Loại phép biến đổi này được sử dụng để tăng cường hình ảnh cho các loại thiết bị hiển thị khác nhau. Gamma của các thiết bị hiển thị khác nhau cũng khác nhau. Ví dụ, gamma của CRT nằm trong khoảng từ 1.8 đến 2.5, có nghĩa là hình ảnh hiển thị trên CRT sẽ tối hơn.

### Biến Đổi Logarit:

- Các phép biến đổi logarit có thể được định nghĩa bằng công thức sau:

- s=clog(r+1)

- Trong đó, 
- s và r là giá trị pixel của ảnh đầu ra và ảnh đầu vào, và c là một hằng số. Giá trị 1 được cộng vào mỗi giá trị pixel của ảnh đầu vào vì nếu có độ sáng pixel bằng 0 trong ảnh, thì log(0) bằng vô cùng. Do đó, 1 được cộng thêm để đảm bảo giá trị tối thiểu ít nhất là 1.

Trong quá trình biến đổi logarit, các pixel tối trong ảnh được mở rộng so với các giá trị pixel cao hơn. Các giá trị pixel cao hơn bị nén trong biến đổi logarit.


### Biến Đổi Âm

- Phép biến đổi tuyến tính thứ hai là biến đổi âm, là phép biến đổi ngược của phép biến đổi đồng nhất. Trong biến đổi âm, mỗi giá trị của ảnh đầu vào được trừ đi L−1 và ánh xạ lên ảnh đầu ra.

- Phép biến đổi này được thực hiện bằng công thức:
- s=(L−1)−r

- Vì vậy, mỗi giá trị được trừ đi 255, khiến các pixel sáng trở nên tối và các pixel tối trở nên sáng. Kết quả là hình ảnh âm.

### Lab Task:

- Áp dụng các kỹ thuật biến đổi sau đây trên hình ảnh được cung cấp và quan sát hình ảnh đầu ra của mỗi phép biến đổi:

- I. Biến Đổi Âm

- II. Biến Đổi Logarit

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

def negative_transformation(image):
    """
    Thực hiện biến đổi âm trên hình ảnh.
    Công thức: s = (L - 1) - r
    """
    L = 256  # Số mức xám cho ảnh 8-bit
    negative_image = (L - 1) - image
    return negative_image

def log_transformation(image):
    """
    Thực hiện biến đổi logarit trên hình ảnh.
    Công thức: s = c * log(1 + r)
    Trong đó, c = 255 / log(1 + max_pixel_value)
    """
    image = image.astype(np.float32) 
    c = 255 / np.log(1 + np.max(image))
    log_image = c * np.log(1 + image)
    log_image = np.uint8(cv2.normalize(log_image, None, 0, 255, cv2.NORM_MINMAX))
    
    return log_image

def display_images(original, negative, log_transformed):
    """
    Hiển thị hình ảnh gốc, hình ảnh sau biến đổi âm và biến đổi logarit.
    """
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(original, cmap='gray')
    plt.title('Hình Ảnh Gốc')
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.imshow(negative, cmap='gray')
    plt.title('Biến Đổi Âm')
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.imshow(log_transformed, cmap='gray')
    plt.title('Biến Đổi Logarit')
    plt.axis('off')

    plt.show()

if __name__ == "__main__":
    path_img = r'D:\Image-Proccessing\pratice\week_1\img\00750.jpeg'

    image = cv2.imread(path_img,0)
    negative_image = negative_transformation(image)
    log_transformed_image = log_transformation(image)
    display_images(image, negative_image, log_transformed_image)
```

### Nhiệm vụ 2: Áp dụng các phép biến đổi sau trên một hình ảnh.

- Phạm vi giá trị pixel và giá trị pixel đầu ra:

- a) Nếu giá trị nhỏ hơn trung bình:
- Giá trị pixel đầu ra = 0
- Nếu giá trị lớn hơn trung bình:
- Giá trị pixel đầu ra = 255
- 
- b) Nếu giá trị nhỏ hơn trung bình:
- Giá trị pixel đầu ra = 255
- Nếu giá trị lớn hơn trung bình:
- Giá trị pixel đầu ra = 0
- 
- c) Nếu giá trị nằm trong khoảng ±20 của trung bình:
- Giá trị pixel đầu ra = 0
- Nếu không:
- Giá trị pixel đầu ra = 255

```python
def apply_transformations(image,mode):
    """
    mode 1: < mean =0 , > mean = 255
    mode 2: < mean = 255 , > mean = 0
    mode 3: Trong khoảng +- 20  = 0, ngoài khoảng +- 20 = 255
    """
    transformed_image = np.zeros_like(image)
    
    if mode == 1:
        transformed_image[image < np.mean(image)] = 0
        transformed_image[image >= np.mean(image)] = 255
    elif mode == 2:
        transformed_image[image < np.mean(image)] = 255
        transformed_image[image >= np.mean(image)] = 0
    elif mode == 3:
        transformed_image[np.abs(image - np.mean(image)) <= 20] = 0
        transformed_image[np.abs(image - np.mean(image)) > 20] = 255
        
    return transformed_image

def display_transformed_images(original, transformed1, transformed2, transformed3):
    """
    Hiển thị hình ảnh gốc và hình ảnh sau biến đổi.
    """
    plt.figure(figsize=(15,10))

    plt.subplot(2,2,1)
    plt.imshow(original, cmap='gray')
    plt.title('Hình Ảnh Gốc')
    plt.axis('off')

    plt.subplot(2,2,2)
    plt.imshow(transformed1, cmap='gray')
    plt.title('Biến Đổi 1')
    plt.axis('off')

    plt.subplot(2,2,3)
    plt.imshow(transformed2, cmap='gray')
    plt.title('Biến Đổi 2')
    plt.axis('off')

    plt.subplot(2,2,4)
    plt.imshow(transformed3, cmap='gray')
    plt.title('Biến Đổi 3')
    plt.axis('off')

    plt.show()

if __name__ == "__main__":
    path_img = r'D:\Image-Proccessing\pratice\week_1\img\00750.jpeg'

    image = cv2.imread(path_img,0)
    transformed_image1 = apply_transformations(image, 1)
    transformed_image2 = apply_transformations(image, 2)
    transformed_image3 = apply_transformations(image, 3)
    display_transformed_images(image, transformed_image1, transformed_image2, transformed_image3)
    
```

### Nhiệm vụ 3: Áp dụng Phép Biến Đổi Luật Lũy Thừa cho các giá trị γ (0.2, 0.5, 1.2 và 1.8)

- Dưới đây là mã Python thực hiện phép biến đổi Luật Lũy Thừa (Power Law Transformation) cho các giá trị của γ là 0.2, 0.5, 1.2 và 1.8. Hãy đảm bảo rằng kiểu dữ liệu được điều chỉnh phù hợp sau mỗi biến đổi.

```python
def power_law_transformation(image, gamma):
    """
    Thực hiện biến đổi luật lũy thừa với giá trị gamma.
    Công thức: s = c * r^gamma
    Trong đó c là hằng số, và r là giá trị pixel.
    """
    # Chuẩn bị ảnh ở định dạng float32 để thực hiện tính toán chính xác
    image = image.astype(np.float32) / 255.0
    
    # Áp dụng phép biến đổi Luật Lũy Thừa
    transformed_image = np.power(image, gamma)
    
    # Chuẩn hóa về thang 0-255 và chuyển về kiểu uint8
    transformed_image = np.uint8(transformed_image * 255)
    
    return transformed_image

def display_images(original, transformed_images, gamma_values):
    """
    Hiển thị các hình ảnh gốc và sau khi áp dụng biến đổi luật lũy thừa.
    """
    plt.figure(figsize=(15, 10))

    plt.subplot(2, 3, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Hình Ảnh Gốc')
    plt.axis('off')

    for i, gamma in enumerate(gamma_values):
        plt.subplot(2, 3, i+2)
        plt.imshow(transformed_images[i], cmap='gray')
        plt.title(f'Gamma = {gamma}')
        plt.axis('off')

    plt.show()

if __name__ == "__main__":
    path_img = r'D:\Image-Proccessing\pratice\week_1\img\00750.jpeg'

    image = cv2.imread(path_img, 0)
    gamma_values = [0.5, 1.0, 1.5, 2.0, 2.5]
    transformed_images = [power_law_transformation(image, gamma) for gamma in gamma_values]
    display_images(image, transformed_images, gamma_values)
```

### Nhiệm Vụ 4: Áp Dụng Cắt Cấp Độ Xám với Giới Hạn Dưới 100 và Giới Hạn Trên 200, Thiết Lập Các Giá Trị Trong Phạm Vi Thành 210

- Dưới đây là mã Python để thực hiện cắt cấp độ xám (Gray Level Slicing) theo yêu cầu:

    - Các pixel có giá trị từ 100 đến 200 sẽ được gán giá trị 210.
    - Các pixel ngoài phạm vi này sẽ giữ nguyên giá trị của chúng.

```python
def gray_level_slicing(image):
    """
    Áp dụng cắt cấp độ xám:
    """
    transformed = np.copy(image)
    rows, cols = image.shape
    
    for i in range(rows):
        for j in range(cols):
            if image[i, j] >= 100 and image[i, j] <= 200:
                transformed[i, j] = 210
            else:
                transformed[i, j] = image[i, j]
    
    return transformed

def display_images(original, transformed):
    """
    Hiển thị hình ảnh gốc và hình ảnh đã áp dụng cắt cấp độ xám.
    """
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(original, cmap='gray')
    plt.title('Hình Ảnh Gốc')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(transformed, cmap='gray')
    plt.title('Hình Ảnh Sau Cắt Cấp Độ Xám')
    plt.axis('off')

    plt.show()

if __name__ == "__main__":
    path_img = r'D:\Image-Proccessing\pratice\week_1\img\00750.jpeg'

    image = cv2.imread(path_img, 0)
    sliced_image = gray_level_slicing(image)
    display_images(image, sliced_image)
```
